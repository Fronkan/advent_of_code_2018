```python
from operator import itemgetter

data = [line for line in open("day4_data.txt")]
data_cleaned = []
for elem in data:
    split = elem.split("]")
    date_time = split[0][1:].split()
    data_cleaned.append({"date": date_time[0], "time": date_time[1],"event": split[1].strip()})

sorted_data = sorted(data_cleaned, key=itemgetter("date","time"))
sorted_data
```

```python
import re
sleep_schedual = {}
for elem in sorted_data:
    gaurd_numbers = re.findall(r'#\d+', elem["event"])
    if gaurd_numbers != []:
        cur_guard = gaurd_numbers[0].strip("#")
        is_asleep = False
        if cur_guard not in sleep_schedual:
            sleep_schedual[cur_guard] = []
        
    if elem["event"] == "falls asleep":
        sleep_time = elem["time"]
        
    elif elem["event"] == "wakes up":
        sleep_schedual[cur_guard].append((sleep_time, elem["time"])) 
        
sleep_schedual
```

```python
sleep_counts = {}
tot_sleep = {}
for guard in sleep_schedual:
    sleep_counts[guard] = [0 for i in range(60)]
    tot_sleep[guard] = 0
    for time in sleep_schedual[guard]:
        t1 = int(time[0].split(":")[1])
        t2 = int(time[1].split(":")[1])
        for t in range(t1,t2):
            sleep_counts[guard][t] += 1
            tot_sleep[guard] += 1
sleep_counts
```
